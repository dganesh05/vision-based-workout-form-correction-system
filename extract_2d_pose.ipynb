```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
```

```python
# Load the pre-trained YOLOv8 pose model
# You can swap 'n' for 's', 'm', 'l', or 'x' for larger models
model = YOLO('yolov8n-pose.pt')

# Fill this with your video path
# Example:
# video_path = '/content/gdrive/Shareddrives/YOUR_SHARED_DRIVE/YOUR_PROJECT_FOLDER/videos/squat_video.mp4'
video_path = '/content/gdrive/Shareddrives/YOUR_SHARED_DRIVE/YOUR_PROJECT_FOLDER/videos/YOUR_VIDEO.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise FileNotFoundError(f'Could not open video: {video_path}')
```

```python
# Preview annotated frames in Colab (no keyboard interrupt in output cell)
frame_idx = 0
max_preview_frames = 30
preview_every_n_frames = 10
shown = 0

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    results = model.track(frame, persist=True)
    annotated_frame = results[0].plot()

    if frame_idx % preview_every_n_frames == 0:
        cv2_imshow(annotated_frame)
        shown += 1
        if shown >= max_preview_frames:
            break

    frame_idx += 1

cap.release()
print('Done. Released video capture.')
```
